```python
# Plot a scatter plot for two numerical columns
plt.scatter(data["Price"], data["Rooms"])
plt.title("Scatter Plot of Price vs Rooms")
plt.xlabel("Price")
plt.ylabel("Rooms")
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python

# Compute the correlation matrix
data_number = data.select_dtypes(include=['number'])
correlation_matrix = data_number.corr()

# Plot the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
